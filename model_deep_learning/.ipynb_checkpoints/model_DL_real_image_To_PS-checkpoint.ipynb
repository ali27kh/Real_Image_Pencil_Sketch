```python
# Importing necessary libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
import tensorflow as tf
from tensorflow.keras import layers, models
from keras.layers import BatchNormalization, Conv2D, Conv2DTranspose, ReLU, Input, LeakyReLU
```

```python
def load_colored_faces(folder_path, target_size=(224, 224)):
    images = []
    labels = []
        
    if os.path.isdir(folder_path):
        # Sort the list of filenames
        filenames = sorted(os.listdir(folder_path))
        for filename in filenames:
            img_path = os.path.join(folder_path, filename)
            
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
                img = cv2.resize(img, target_size)  # Resize to target size
                
                img_array = img / 255.0  # Normalize pixel values to be between 0 and 1
                images.append(img_array)
                labels.append(0)  # Assigning a label (currently set to 0, modify as needed)
            
    return np.array(images), np.array(labels)

def load_pencil_faces(folder_path, target_size=(224, 224)):
    images = []
    labels = []
        
    if os.path.isdir(folder_path):
        # Sort the list of filenames
        filenames = sorted(os.listdir(folder_path))
        for filename in filenames:
            img_path = os.path.join(folder_path, filename)
            
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Read as grayscale
            if img is not None:
                img = cv2.resize(img, target_size)  # Resize to target size
                
                img_array = img / 255.0  # Normalize pixel values to be between 0 and 1
                images.append(img_array)
                labels.append(1)  # Assigning a label (currently set to 1, modify as needed)
                
    return np.array(images), np.array(labels)
```

```python
color_faces,_ = load_colored_faces('data/RGB_Faces')
pencil_faces,_ = load_pencil_faces('data//Pencil_Faces')
```

```python
print(color_faces.shape,pencil_faces.shape)
```

```python
# Display the first 4 images from the dataset
plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(1, 5, i + 1)
    plt.imshow(color_faces[i])
    plt.title(f"Colored Face {i + 1}")
    plt.axis('off')

plt.show()

plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(1, 5, i + 1)
    plt.imshow(pencil_faces[i],cmap='gray')
    plt.title(f"Pencil Face {i + 1}")
    plt.axis('off')

plt.show()
```

```python
# Expand the dimensions of pencil_faces to make it 3D
pencil_face_three_dim = np.expand_dims(pencil_faces, axis=-1)

# This is done to match the format of the colored faces (3 channels for RGB)
pencil_face_three_dim = np.tile(pencil_face_three_dim, (1, 1, 3))
```

```python
pencil_faces = pencil_face_three_dim
```

```python
print(pencil_faces.shape)
print(color_faces.shape)
```

## Split data (train/test):

```python
x_train_color,x_test_color,x_train_pencil,x_test_pencil=train_test_split(color_faces,pencil_faces,test_size=0.2,random_state=100,shuffle= True)
```

```python
print(x_train_color.shape,x_train_pencil.shape)
print(x_test_color.shape,x_test_pencil.shape)
```

```python
def build_pencil_sketch_model(input_shape=(224, 224, 3)):
    model = Sequential()

    # Encoder
    model.add(Conv2D(64, (3, 3), strides=(2, 2), padding='same', input_shape=input_shape))
    model.add(ReLU())
    model.add(BatchNormalization())
    model.add(Conv2D(128, (3, 3), strides=(2, 2), padding='same'))
    model.add(ReLU())
    model.add(BatchNormalization())
    model.add(Conv2D(256, (3, 3), strides=(2, 2), padding='same'))
    model.add(ReLU())
    model.add(BatchNormalization())

    # Decoder
    model.add(Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same'))
    model.add(ReLU())
    model.add(BatchNormalization())
    model.add(Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same'))
    model.add(ReLU())
    model.add(BatchNormalization())
    model.add(Conv2DTranspose(3, (3, 3), strides=(2, 2), padding='same', activation='sigmoid'))

    return model
```

```python
model.summary()
```

```python
model = build_pencil_sketch_model()
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

history = model.fit(x_train_color, x_train_pencil, epochs=20, batch_size=16,
                    validation_data=(x_test_color, x_test_pencil))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Loss (Train)', color='blue')
plt.plot(history.history['val_loss'], label='Loss (Validation)', color='red')
plt.title('Courbe de la perte pendant l\'entranement')
plt.xlabel('poques')
plt.ylabel('Perte')
plt.legend()
plt.show()
```

```python
predicted_sketches = model.predict(x_test_color)

plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(2, 4, i + 1)
    plt.imshow(x_test_color[i])
    plt.title(f"Image relle {i + 1}")
    plt.axis('off')
    plt.subplot(2, 4, i + 5)
    plt.imshow(predicted_sketches[i])
    plt.title(f"Sketch gnr {i + 1}")
    plt.axis('off')

plt.show()
```

```python
# Enregistrer le modle au format .keras (.h5)
model.save('pencil_sketch_model.keras')
```
